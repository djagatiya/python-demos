```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.describe()
```

```python
for col_name in df.select_dtypes(include=np.number).columns:
    fit, ax = plt.subplots(2,1, figsize=(6, 6))
    sns.histplot(data=df, x=col_name, ax=ax[0])
    sns.boxplot(data=df, x=col_name, ax=ax[1])
    
    # mean
    _mean = df[col_name].mean()
    ax[0].axvline(x=_mean, label='mean', color='red')
    
    # std
    _std = df[col_name].std()
    
    ax[0].axvspan(_mean - _std,  _mean + _std, alpha=0.5)
    
    ax[0].legend()
    
    print(" Mean:", _mean, "\n", "Std:", _std, "\n", "Median:", df[col_name].median())
    
    plt.tight_layout()
    plt.show()
```
