```python
df = pd.read_csv("../input/10000-most-common-passwords/common_passwords.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
num_cols = ['length', 'num_chars', 'num_digits', 'num_upper', 'num_lower', 'num_special', 'num_vowels', 'num_syllables']
```

```python
for name in num_cols:
    plt.figure(figsize=(6,2))
    sns.histplot(data=df, x=name)
```

```python
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.tight_layout()      
```

```python
count_series_ls = [df.groupby(name).size().rename(name) for name in num_cols] 
pd.concat(count_series_ls, axis=1).style.background_gradient(cmap='coolwarm', axis=0)
```
